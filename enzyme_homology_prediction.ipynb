```python
# @title
import pandas as pd

df = pd.read_csv("sample_data/california_housing_test.csv")

print(df.head())
```

```python
import plotly.graph_objects as go

## 3d scatter plot of longitude, latitude, and population

fig = go.Figure(
    data = [
      go.Scatter3d(
          x = df['longitude'],
          y = df['latitude'],
          z = df['population'],
          mode = 'markers',
          marker = dict(
              size = 3,
              opacity= 0.5
          )
      )
    ]
)
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
fig.show()
```
